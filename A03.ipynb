```python
import pandas as pd
import statsmodels.api as sm


df = pd.read_excel("Aguas.xlsx")
```

## Convertir a datos bool


```python
df['Tamaño_Bool'] = df['Tamaño'].map({'Grande': 1, 'Chica': 0})
df['Hielo_Bool'] = df['Hielo'].map({'Con': 1, 'Sin': 0})
df['Kahlua'] = df['Sabor'].map({'Kahlua': 1, 'Horchata': 0})
df['Horchata'] = df['Sabor'].map({'Horchata': 1, 'Kahlua': 0})
print(df)
```

## Regresion

```python
y = df['Score']
x = df[['Tamaño_Bool', 'Hielo_Bool', 'Kahlua', 'Horchata']]

x_const = sm.add_constant(x)
modelo = sm.OLS(y, x_const).fit()

print(modelo.summary())
```

```python
df['Tamaño_Bool'] = df['Tamaño'].map({'Grande': -1, 'Chica': 1})
df['Hielo_Bool'] = df['Hielo'].map({'Con': 1, 'Sin': -1})
df['Kahlua'] = df['Sabor'].map({'Kahlua': 1, 'Horchata': 0})
df['Horchata'] = df['Sabor'].map({'Horchata': 1, 'Kahlua': 0})
print(df)
```

```python
import statsmodels.api as sm

X_bin = sm.add_constant(df[['Tamaño_Bool','Hielo_Bool','Kahlua','Horchata']])
modelo_bin = sm.OLS(y, X_bin).fit()
print(modelo_bin.summary())
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Prepare data
X = salarios_minimos_df[['Periodo']]
y = salarios_minimos_df['Salario']

# Create range for smooth plotting
X_plot = np.linspace(X.min(), X.max(), 500).reshape(-1, 1)

# Plot setup
plt.figure(figsize=(12, 8))
plt.scatter(X, y, label='Datos Originales', color='blue', alpha=0.6)

# Train and plot models
for degree in [1, 3, 5, 7]:
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X, y)
    plt.plot(X_plot, model.predict(X_plot), label=f'Polinomio orden {degree}')

# Add plot decorations
plt.title('Regresión Polinomial: Salarios Mínimos vs. Periodo')
plt.xlabel('Periodo')
plt.ylabel('Salario Mínimo')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Prepare data
X = bateria_tableta_df['carga'].values.reshape(-1, 1)
y = bateria_tableta_df['duracion'].values

# Create X values for smooth plotting
X_plot = np.linspace(X.min(), X.max(), 500).reshape(-1, 1)

# Define polynomial orders
degrees = [1, 3, 5, 7]

# Plot setup
plt.figure(figsize=(12, 8))
plt.scatter(X, y, label='Datos Originales', color='red', alpha=0.6)

# Train and plot models in one loop
for degree in degrees:
    # Create polynomial regression model using pipeline
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X, y)

    # Predict and plot
    y_pred = model.predict(X_plot)
    plt.plot(X_plot, y_pred, label=f'Polinomio orden {degree}')

# Add plot decorations
plt.title('Regresión Polinomial: Carga vs. Duración de Batería')
plt.xlabel('Carga de Batería')
plt.ylabel('Duración de Batería')
plt.legend()
plt.grid(True)
plt.show()
```
